```python
# PCA with breast cancer
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# importing pre-loaded datasets
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['target_names'])
```

```python
print(cancer['DESCR'])
```

```python
print(cancer['feature_names'])
```

```python
# Creating a column with data from 'feature_names'.
# Adding all data from 'data' and adding them to columns in 'feature_names'
df = pd.DataFrame(cancer['data'], columns = cancer['feature_names'])
df.head(10)
```

```python
df.shape
```

```python
# Reducing from 30 features to 2 feature
# Standard Scaling (Standard_deviation=1 and mean=0)
# can also use MinMaxscaler (convert all values between [0-1])
# difference between fit and fit_transform 
# (link: https://datascience.stackexchange.com/questions/12321/whats-the-difference-between-fit-and-fit-transform-in-scikit-learn-models#:~:text=In%20summary%2C%20fit%20performs%20the,in%20one%20possibly%20optimized%20step.)
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
scaled_data = sc.fit_transform(df)
```

```python
## Standard Scaling (Standard_deviation=1 and mean=0)
scaled_data
```

```python
# PCA
from sklearn.decomposition import PCA

# Reducing to 2 dimensions
pca = PCA(n_components=2)
```

```python
x_pca = pca.fit_transform(scaled_data)
```

```python
# dimensions of scaled data (30)
scaled_data.shape
```

```python
# Dimensions after applying PCA (2)
x_pca.shape
```

```python
# creating a figure with matplotlib
# for creating three dimensional scatterplot
# (blog: https://drzinph.com/pca-visualized-with-3d-scatter-plots/)
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1], c = cancer['target'])
plt.xlabel('First principal components')
plt.ylabel('Second principal components')
plt.title('Scatterplot of 2-D PCA', weight='bold')
```
